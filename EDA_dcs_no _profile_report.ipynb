If we look at the heading of the first column it is named __Unnamed 0__ but since it is representing months of years so we rename it as __Year__ and convert into into YYYY-MM_DD format 

```python
#Renaming

df_cpi_monthly.rename(columns = {'Unnamed: 0':'Year'} , inplace = True)
```

```python
df_cpi_monthly.dtypes
```

```python
# conversion into YYYY-MM-DD Format.

# For YYYY we have selected from 0th index to 4th index    
l1 = [ ]
for i in range(len(df_cpi_monthly)):
    l1.append(df_cpi_monthly['Year'][i][0:4])    

# For MM format we have selected 5th and 6th index as they represent the month number.  
l2 = [ ]
for i in range(len(df_cpi_monthly)):
    l2.append(df_cpi_monthly['Year'][i][5:7])    

# Now to add YYYY and MM together we create another list
l3 = [ ]
for i in range(len(l1)):
    l3.append(l2[i]+'-'+l1[i])
```

```python
l3
```

```python
def prepend(list, str):
 
    # Using format()
    str += '{0}'
    list = [str.format(i) for i in list]
    return(list)
a = '01-'
l3 = prepend(l3,a)
```

```python
l3
```

```python
from datetime import datetime

l3 = [datetime.strptime(x,'%d-%m-%Y') for x in l3]
```

```python
# Find the name of the column by index
n = df_cpi_monthly.columns[0]

# Drop that column
df_cpi_monthly.drop(n, axis = 1, inplace = True)

# Put whatever series you want in its place

df_cpi_monthly.insert(0,'Year',l3)
```

Similarly linke CPI dataframes here also we need to convert Period into Year column and change as 5__YYYY-MM-DD__ Format in __monthly dataframe__ . Also observe that the name of some countries are misspelt such as __Argentina__ with __Argenti0__ , __Burkina Faso__ with __Burki0 Faso__ etc. so we need to make that also correct.

```python
df_exchange_monthly.head()
```

```python
# Renaming Period Column and Changing it's format 

df_exchange_monthly.rename(columns = {'Period':'Year'} , inplace = True)
```

```python
# For YYYY we have selected from 0th index to 4th index    
l1 = [ ]
for i in range(len(df_exchange_monthly)):
    l1.append(df_exchange_monthly['Year'][i][0:4])    

# For MM format we have selected 5th and 6th index as they represent the month number.  
l2 = [ ]
for i in range(len(df_exchange_monthly)):
    l2.append(df_exchange_monthly['Year'][i][5:7])    

# Now to add YYYY and MM together we create another list
l3 = [ ]
for i in range(len(l1)):
    l3.append(l2[i]+'-'+l1[i])
    
def prepend(list, str):
 
    # Using format()
    str += '{0}'
    list = [str.format(i) for i in list]
    return(list)
a = '01-'
l3 = prepend(l3,a)

from datetime import datetime

l3 = [datetime.strptime(x,'%d-%m-%Y') for x in l3]


# Find the name of the column by index
n = df_exchange_monthly.columns[0]

# Drop that column
df_exchange_monthly.drop(n, axis = 1, inplace = True)

# Put whatever series you want in its place

df_exchange_monthly.insert(0,'Year',l3)
```

```python
df_exchange_monthly.head()
```

```python
df_exchange_monthly.dtypes
```

```python
# Now renaming Columns with misspelt name :

# In the column names at place of 'na' it is repolce with '0' so we need to change that 

l  = list(df_exchange_monthly.columns)   # first we create a list with all columns 
```

```python
# now we create another list that will contain correct column  names 

new_list = [ ]
for item in l:
    if '0' in item:
        new_list.append(str(item).replace('0' , 'na'))
    else:
        new_list.append(str(item))
```

```python
new_list
```

```python
# Now we replace the column names: 

df_exchange_monthly.columns = new_list
```

```python
df_exchange_monthly.head()
```
