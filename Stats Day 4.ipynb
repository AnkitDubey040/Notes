#### Understanding CLT

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Firstly we will define a random population of size 1 million
pop_size = 1000000
# now we randomly get the random population
population = np.random.rand(1000000)
```

```python
# the population is 
population
```

```python
# now we will get a sample of size 10000 from the population
sample_size = 10000
# now randomly have means of these samples 
sample_means = np.random.rand(sample_size)
```

```python
sample_means
```

```python
# now we find mean of those samples
c = np.random.rand(sample_size)
for i in range(0,sample_size):
    c=np.random.randint(1,pop_size,sample_size)
    sample_means[i] = np.mean(c)
```

```python
np.random.randint(1,pop_size,sample_size)
```

```python
sample_means.shape
```

```python
# Now we create histogram and bell curve for the distribution of means 
plt.subplot(1,2,1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_means,bins=int(180/5),hist = True,kde = False)
plt.title('Histogram of Sample mea',fontsize=20)
plt.xlabel('Sample mean',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.subplot(1,2,2)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_means,hist = False,kde = True)
plt.title('Density of Sample mean',fontsize=20)
plt.xlabel('Sample mean',fontsize=20)
plt.ylabel('Density',fontsize=20)
plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
```
