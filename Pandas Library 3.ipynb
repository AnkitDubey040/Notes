```python
df_sales = pd.read_csv("E:\DATA SCIENCE\data fsds_\sales_data_final (1).csv")
```

```python
# we want new column where we want to store derived data in that column
#say we want to find from profit column whether profit positive or negative and print out it as column
```

```python
def profit_flag(a):
    if a>= 0 :
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(213)
```

```python
profit_flag(-231)
```

```python
# now we want to pass data of profit column as parameter 
'''
for that we use APPLY
'''
```

```python
'''we create flag_profit column '''
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
# now see the table for new column 
df_sales.head()
```

```python
# now we want to create another column where we store length of customer name
```

```python
df_sales['len_customer_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
# now we want to find quantity sold is high , medium or low
def quantity_flag(a) : 
    if  a < 10 :
        return 'low'
    elif a >10 and a <20 :
        return 'medium'
    else :
        return 'high'
```

```python
df_sales['flag_quantity']= df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head(20)
```

```python
# using LAMBDA func and Apply
# to find square of quantity
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a**2)
```
