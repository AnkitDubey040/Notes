##### 5 . which id is one of the active id that you have in whole dataset 

```python
# let's say we find activeness by total steps so find most active ID as:
df.groupby('Id')['TotalSteps'].sum().sort_values(ascending=False).head(1)
```

```python
# If we consider activeness by calorie then most active is given by: 
df.groupby('Id')['Calories'].sum().sort_values(ascending=False).head(1)
```

##### 6 . how many of them have not logged there activity find out in terms of number of ids

```python
# This is the no. of users that have logged in at least once :
df1 = df[df['LoggedActivitiesDistance'] >0]
```

```python
# the number of unique user are :
df1.groupby('Id')['LoggedActivitiesDistance'].sum()
```

```python
len(df1.groupby('Id')['LoggedActivitiesDistance'].sum())
```

```python
# now to find no of users who have not logged activity we subtract No of uniques user with no of active users:
len(df['Id'].unique()) - len(df1.groupby('Id')['LoggedActivitiesDistance'].sum())
```

```python
# so the users that have not logged activity are 29
```
