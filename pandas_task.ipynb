###### Ans13:
Total contributation hour for each and every agents weekly basis 

```python
# here we need to use Agent logging csv file 
df10 = pd.read_csv("E:\DATA SCIENCE\data fsds_\AgentLogingReport.csv" , index_col = 'SL No')
```

```python
df10
```

```python
df10.dtypes
```

```python
l = df10['Duration'].str.split(":")
df10['hours'] = (l.str[0].astype(int) * 3600 + l.str[1].astype(int) * 60 + l.str[2].astype(int)) / 3600
df10['Date'] = pd.to_datetime(df10['Date'])
df10['Week'] = df10['Date'].dt.week
```

```python
df11 = df10.groupby(['Week','Agent'])['hours'].sum()
df11 = df11.to_frame(name='hour contributed per week')
df11
```

###### Ans14:
total percentage of active hour for a month 


```python
df12 = ((df10.groupby('Agent')['hours'].sum()) / (30*24)) * 100
df12 = df12.to_frame('Active hours percentage')
df12
```
