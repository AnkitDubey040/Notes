###### 11 . Gorup by region and find out which region is more profitable 


```python
Orders_df.groupby('Region')['Profit'].mean().sort_values(ascending = False).head(1)
```

```python
# Cetral region has profit of 179.312027 per order
```

```python
# if we want total profit by sum then: 
Orders_df.groupby('Region')['Profit'].sum().sort_values(ascending = False).head(1)
```

```python
# By sum total California has highest profit
```

##### 12 . Try to find out overalll in which country we are giving more discount 


```python
# Per Order
Orders_df.groupby('State or Province')['Discount'].mean().sort_values(ascending = False).head(1)
```

```python
# By sum total
Orders_df.groupby('State or Province')['Discount'].sum().sort_values(ascending = False).head(1)
```

##### 14 . which customer segement is more profitalble find it out . 

```python
# By average
Orders_df.groupby('Customer Segment')['Profit'].mean().sort_values(ascending = False).head(1)
```

```python
# By  Sum
Orders_df.groupby('Customer Segment')['Profit'].sum().sort_values(ascending = False).head(1)
```

##### 15 . try to find out the 10th most loss making product  sub catagory . 

```python
#By mean 
Orders_df.groupby('Product Sub-Category')['Profit'].mean().sort_values(ascending = True).index[9]
```

```python
# By sum 
Orders_df.groupby('Product Sub-Category')['Profit'].sum().sort_values(ascending = True).index[9]
```
